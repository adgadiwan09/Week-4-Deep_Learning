```
import tensorflow as tf
import numpy as np
import tensorflow.keras as keras
import pandas as pd
```

```
df= pd.read_csv('diabetes.csv')
df.head() 
```

```
#input will be from columns 1-8 i.e. everything except Outcome which is the only output.
x=df.iloc[:,0:8].values     #input
y=df.iloc[:,8].values       #output
```

```
#defining the keras model
model= tf.keras.Sequential()
model.add(keras.layers.Dense(12,input_dim =8, activation='relu' ))  #hiddenlayer
model.add(keras.layers.Dense(8, activation='relu' ))                #hiddenlayer
model.add(keras.layers.Dense(1, activation='sigmoid' ))             #outputlayer
```

```
model.compile(optimizer='adam' ,loss='binary_crossentropy' ,metrics=['accuracy'])
```

```
model.fit(x,y,epochs=300,batch_size=10)
```

```
predictions = model.predict_classes(x)
```

```
predictions
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y,predictions)
```

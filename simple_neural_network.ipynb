```
import tensorflow as tf
import tensorflow.keras as keras  # HIGH Level API
import numpy as np
```

```
# Detect the pattern between input and output
```

```
x = np.array([1,2,3,4,5]) # Input
y = np.array([-100,100,400,500,800]) #Output
```

```
#ypred taken randomly just to see how to calculate error
#ypred = np.array([-1,0,2,4,7])
#error = np.square(np.subtract(y,ypred)).mean()
#error
```

```
model = tf.keras.Sequential([keras.layers.Dense(units = 1,input_shape=[1])])
```

```
model.compile(optimizer = 'sgd',loss = 'mean_squared_error')
model.fit(x,y,epochs=500) #In ML we dont have epochs
```

```
model.predict(x)
```
